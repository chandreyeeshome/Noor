```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/Book2 (3).csv").drop("Analysis", axis=1)
```

```python
df.head()
```

```python
df.columns
```

```python
TARGETS = ['Anxiety', 'Independence', 'Extroversion', 'Tough_Mindness', 'Self_Control']
```

```python
X = df.drop(TARGETS, axis=1)
y = df[TARGETS]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=16)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.multioutput import MultiOutputClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
models = [DecisionTreeClassifier(), RandomForestClassifier(), SVC()]
```

```python
for model in models:
    multi_model = MultiOutputClassifier(model)
    multi_model.fit(X_train, y_train)
    acc = multi_model.score(X_test, y_test)
    print(f"Model: {model} \t Accuracy: {acc}")
```
